```python
import matplotlib.pyplot as plt # plt är namn på matplotlibs plot funktion
import numpy as np      # numpy används för matematiska operationer och arrays


# data type_ arrays
```

```python
# generate (skapa) data x array med 100 tal mellan -5 och 5
x = np.linspace (-5, 5, 100)
x
```

```python
# generate data f / # Beräkna sinusvärdet för varje x och sparar det i f.
f = np.sin(x)
f
```

The first approach - functional/matlab style  (النهج الأول (الطريقة أو الأسلوب))

```python
plt.figure()
plt.plot (x, f)     # ritar en linje som visar f mot x
plt.title ("sin(x)")
plt.xlabel ("x")
plt.ylabel("f")
```

The second approach - OOP

```python
fig = plt.figure() # Creat a new figure
ax = plt.axes() # Creat a new axis on the figure
ax.plot (x, f) #creat a plot on the axe
ax.set (xlim = (-10, 10), xlabel ="x", ylabel ="y") # continue modify the axe object in different aspects
```
